```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from eval_nondeterministic import evaluate
```

```python
n_users = [10, 20, 30]
ratio = [0.1, 0.3, 0.5, 1]
recipes_per_cuisine = 50
weight_rating_features = 0.5
top_k = [1, 2, 3, 5]
```

```python
d = dict([(r,[evaluate(u, recipes_per_cuisine, r, r, 0.5) for u in n_users]) for r in ratio])
df = pd.DataFrame(d, index=n_users)
```

```python
sns.set()

for r in ratio:
    df_abs = pd.DataFrame({'k':[] , 'absolute accuracy':[], 'users':[]})
    df_rel = pd.DataFrame({'k':[] , 'relative accuracy':[], 'users':[]})
    for u in n_users:
        absolute, relative = evaluate(u, recipes_per_cuisine, r, r, 0.5)
        d_abs = {'k':[] , 'absolute accuracy':[], 'users':[]}
        d_rel = {'k':[], 'relative accuracy':[], 'users':[]}
        for (k, (val_a, val_r)) in zip(top_k, zip(absolute, relative)):
            d_abs['k'].append('k = ' + str(k))
            d_rel['k'].append('k = ' + str(k))
            d_abs['users'].append(u)
            d_rel['users'].append(u)
            d_abs['absolute accuracy'].append(val_a)
            d_rel['relative accuracy'].append(val_r)
        df_abs = df_abs.append(pd.DataFrame(d_abs))
        df_rel = df_rel.append(pd.DataFrame(d_rel))

    sns.lineplot(data=df_abs, x='users', y='absolute accuracy',hue='k', color="g")
    plt.xticks(n_users)
    plt.title('absolute accuracy for {} recipes per cuisine and ratings for {}% of all recipes'.format(recipes_per_cuisine, r*100))
    plt.show()
    plt.savefig('plot_users_ratio/' + 'ratio_{}_absolute.png'.format(r))
    
    sns_plot = sns.lineplot(data=df_rel, x='users', y='relative accuracy',hue='k', color="r")
    plt.xticks(n_users)
    plt.title('relative accuracy for {} recipes per cuisine and ratings for {}% of all recipes'.format(recipes_per_cuisine, r*100))
    plt.savefig('plot_users_ratio/' + 'ratio_{}_relative.png'.format(r))
    plt.show()
```
