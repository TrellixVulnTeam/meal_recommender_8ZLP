```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from eval_nondeterministic import evaluate
```

```python
n_users = [10, 20, 30]
ratio = [0.1, 0.3, 0.5, 1]
recipes_per_cuisine = 50
weight_rating_features = 0.5
top_k = [1, 2, 3, 5]
```

```python
sns.set()

for r in ratio:
    df_abs = pd.DataFrame({'k':[] , 'absolute score':[], 'users':[]})
    df_rel = pd.DataFrame({'k':[] , 'relative score':[], 'users':[]})
    for u in n_users:
        absolute, relative = evaluate(u, recipes_per_cuisine, r, r, 0.5)
        d_abs = {'k':[] , 'absolute score':[], 'users':[]}
        d_rel = {'k':[], 'relative score':[], 'users':[]}
        for (k, (val_a, val_r)) in zip(top_k, zip(absolute, relative)):
            d_abs['k'].append('k = ' + str(k))
            d_rel['k'].append('k = ' + str(k))
            d_abs['users'].append(u)
            d_rel['users'].append(u)
            d_abs['absolute score'].append(val_a)
            d_rel['relative score'].append(val_r)
        df_abs = df_abs.append(pd.DataFrame(d_abs))
        df_rel = df_rel.append(pd.DataFrame(d_rel))

    sns.lineplot(data=df_abs, x='users', y='absolute score',hue='k', color="g")
    plt.xticks(n_users)
    plt.title('{} recipes per cuisine and ratings for {}% of all recipes'.format(recipes_per_cuisine, r*100))
    plt.savefig('plot_vary_users/' + 'ratio_{}_absolute.png'.format(r))
    plt.show()

    sns_plot = sns.lineplot(data=df_rel, x='users', y='relative score',hue='k', color="r")
    plt.xticks(n_users)
    plt.title('{} recipes per cuisine and ratings for {}% of all recipes'.format(recipes_per_cuisine, r*100))
    plt.savefig('plot_vary_users/' + 'ratio_{}_relative.png'.format(r))
    plt.show()
```
