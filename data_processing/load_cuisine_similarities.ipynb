```python
import pandas as pd
import pickle
```

```python
df = pd.read_csv('../data/cuisine_similarities/cuisine_similarities.csv')
```

```python
df
```

```python
cuisine_similarities = {}

for c in df.cuisine:
    cuisine_similarities[c] = {}


for i, c1 in enumerate(df.cuisine):
        for c2 in df.cuisine:
            # get rid of nan s
            if df.loc[i][c2] >= 0:
                cuisine_similarities[c1][c2] = df.loc[i][c2]
                cuisine_similarities[c2][c1] = df.loc[i][c2]
```

```python
with open('../data/cuisine_similarities/cuisine_similarities.obj', 'wb') as filehandler:
            pickle.dump(cuisine_similarities, filehandler)
```

```python
# to load:
with open('../data/cuisine_similarities/cuisine_similarities.obj', 'rb') as file_sim:
                cuisine_similarities = pickle.load(file_sim)
```
